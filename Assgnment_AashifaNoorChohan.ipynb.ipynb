## Q1. Arithmetic and Assignment Operators

```python
# Define two numeric variables
a = 15
b = 4

# ── Arithmetic Operators ──────────────────────────────────────────────────────
print("===== Arithmetic Operators =====")
print(f"a = {a}, b = {b}")
print(f"Addition       : a + b  = {a + b}")
print(f"Subtraction    : a - b  = {a - b}")
print(f"Multiplication : a * b  = {a * b}")
print(f"Division       : a / b  = {a / b}")
print(f"Floor Division : a // b = {a // b}")
print(f"Modulus        : a % b  = {a % b}")
print(f"Exponentiation : a ** b = {a ** b}")

# ── Assignment Operators ──────────────────────────────────────────────────────
print("\n===== Assignment Operators =====")
a = 15                          # reset a
print(f"Initial a = {a}")

a += 5                          # a = a + 5
print(f"After a += 5  → a = {a}")

a -= 3                          # a = a - 3
print(f"After a -= 3  → a = {a}")

a *= 2                          # a = a * 2
print(f"After a *= 2  → a = {a}")

a /= 4                          # a = a / 4
print(f"After a /= 4  → a = {a}")

a //= 2                         # a = a // 2
print(f"After a //= 2 → a = {a}")

a %= 3                          # a = a % 3
print(f"After a %= 3  → a = {a}")

a **= 3                         # a = a ** 3
print(f"After a **= 3 → a = {a}")
```

## Q2. Comparison and Logical Operators

```python
# Take two integer inputs from the user
x = int(input("Enter first integer : "))
y = int(input("Enter second integer: "))

# ── Comparison Operators ──────────────────────────────────────────────────────
print("\n===== Comparison Operators =====")
print(f"x = {x}, y = {y}")
print(f"x == y  : {x == y}")
print(f"x != y  : {x != y}")
print(f"x >  y  : {x > y}")
print(f"x <  y  : {x < y}")
print(f"x >= y  : {x >= y}")
print(f"x <= y  : {x <= y}")

# ── Logical Operators ─────────────────────────────────────────────────────────
print("\n===== Logical Operators =====")
# 'and' → True only when BOTH conditions are True
print(f"(x > 0) and (y > 0)  : {(x > 0) and (y > 0)}")
# 'or'  → True when AT LEAST ONE condition is True
print(f"(x > 0) or  (y < 0)  : {(x > 0) or (y < 0)}")
# 'not' → inverts the Boolean value
print(f"not (x == y)          : {not (x == y)}")
```

## Q3. Bitwise Operators

```python
# Define two integers for bitwise demonstration
p = 12   # binary: 1100
q = 10   # binary: 1010

print("===== Bitwise Operators =====")
print(f"p = {p}  (binary: {bin(p)})")
print(f"q = {q}  (binary: {bin(q)})")
print()

# AND – bits that are 1 in BOTH numbers
print(f"p & q  (AND)         = {p & q}   → binary: {bin(p & q)}")

# OR  – bits that are 1 in EITHER number
print(f"p | q  (OR)          = {p | q}   → binary: {bin(p | q)}")

# XOR – bits that are 1 in ONE number but not both
print(f"p ^ q  (XOR)         = {p ^ q}    → binary: {bin(p ^ q)}")

# NOT – flips all bits (gives -(n+1) in Python due to two's complement)
print(f"~p     (NOT)         = {~p}")

# Left Shift – shifts bits left by n positions (multiplies by 2^n)
print(f"p << 2 (Left Shift)  = {p << 2}   → binary: {bin(p << 2)}")

# Right Shift – shifts bits right by n positions (divides by 2^n)
print(f"p >> 2 (Right Shift) = {p >> 2}    → binary: {bin(p >> 2)}")
```

## Q4. Identity and Membership Operators

```python
# Create a list of numbers
numbers = [10, 20, 30, 40, 50]

# ── Identity Operators ────────────────────────────────────────────────────────
print("===== Identity Operators =====")

list1 = numbers          # list1 points to the SAME object as numbers
list2 = [10, 20, 30, 40, 50]  # list2 is a NEW object with same values

# 'is' → True when both variables reference the exact same object in memory
print(f"list1 is numbers      : {list1 is numbers}")
# 'is not' → True when they are different objects
print(f"list2 is not numbers  : {list2 is not numbers}")
print(f"id(numbers) = {id(numbers)}, id(list1) = {id(list1)}, id(list2) = {id(list2)}")

# ── Membership Operators ──────────────────────────────────────────────────────
print("\n===== Membership Operators =====")
print(f"numbers = {numbers}")

# 'in' → True if the value exists inside the list
print(f"30 in numbers         : {30 in numbers}")
print(f"99 in numbers         : {99 in numbers}")

# 'not in' → True if the value does NOT exist inside the list
print(f"99 not in numbers     : {99 not in numbers}")
print(f"10 not in numbers     : {10 not in numbers}")
```

## Q5. String Operations

```python
# Create a string variable with a sentence
sentence = "Data Science is the future of technology and innovation"

print("Original string:")
print(sentence)
print()

# 1. Length of the string
print(f"1. Length of string      : {len(sentence)}")

# 2. Convert to uppercase and lowercase
print(f"2. Uppercase             : {sentence.upper()}")
print(f"   Lowercase             : {sentence.lower()}")

# 3. Find a specific word using the 'in' operator
word_to_find = "future"
print(f"3. Is '{word_to_find}' in sentence? : {'future' in sentence}")

# 4. Replace a word using replace()
new_sentence = sentence.replace("future", "backbone")
print(f"4. After replace()       : {new_sentence}")

# 5. Slice a portion of the string  (characters 0–11)
sliced = sentence[0:12]
print(f"5. Sliced [0:12]         : {sliced}")

# 6. Reverse the string using slicing  (step = -1)
reversed_str = sentence[::-1]
print(f"6. Reversed string       : {reversed_str}")
```
